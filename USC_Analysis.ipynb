### Feature Manipulation

```python
def time_slot(df):

    if (df['Hour'] <=6):
        return 'Slot 1'
    elif (df['Hour'] >= 7 and df['Hour'] <= 12):
        return 'Slot 2'
    elif (df['Hour'] >= 13 and df['Hour'] <= 18):
        return 'Slot 3'
    elif (df['Hour'] >= 19):
        return 'Slot 4'
    
def week_number(df):

    if (df['Date'] >= 1 and df['Date'] <= 7):
        return '1st Week'
    elif (df['Date'] >= 8 and df['Date'] <= 14):
        return '2nd Week'
    elif (df['Date'] >= 15 and df['Date'] <= 21):
        return '3rd Week'
    elif (df['Date'] >= 22 and df['Date'] <= 28):
        return '4th Week'
    else:
        return '5th Week'

def week_slot(df):

    if ( df['Weekday'] <= 4 ):
        return 'Weekday'
    elif (df['Weekday'] >= 5):
        return 'Weekend'
    
def week_day(df):

    if ( df['Week'] == 0 ):
        return 'Monday'
    elif (df['Week'] == 1):
        return 'Tuesday'
    elif (df['Week'] == 2):
        return 'Wednesday'
    elif (df['Week'] == 3):
        return 'Thursday'
    elif (df['Week'] == 4):
        return 'Friday'
    elif (df['Week'] == 5):
        return 'Saturday'
    elif (df['Week'] == 6):
        return 'Sunday'
    
def season(df):

    if ( df['Month'] in ( 'Jan' or 'Feb' or 'Mar' or 'Apr' ) ):
        return 'Spring'
    elif ( df['Month'] in ( 'May' or 'Jun' or 'Jul' )):
        return 'Summer'
    elif ( df['Month'] in ( 'Aug' or 'Sep' or 'Oct' or 'Nov' or 'Dec')):
        return 'Fall'

# Drop Missing Values (NA)
df2012.dropna(subset=['DateOccuredFrom','TimeOccuredFrom','DateOccuredTo','TimeOccuredTo','datereported','Timereported'], inplace=True)
df2013.dropna(subset=['DateOccuredFrom','TimeOccuredFrom','DateOccuredTo','TimeOccuredTo','datereported','Timereported'], inplace=True)
df2014.dropna(subset=['DateOccuredFrom','TimeOccuredFrom','DateOccuredTo','TimeOccuredTo','datereported','Timereported'], inplace=True)
df2015.dropna(subset=['DateOccuredFrom','TimeOccuredFrom','DateOccuredTo','TimeOccuredTo','datereported','Timereported'], inplace=True)
df2016.dropna(subset=['DateOccuredFrom','TimeOccuredFrom','DateOccuredTo','TimeOccuredTo','datereported','Timereported'], inplace=True)
df2017.dropna(subset=['DateOccuredFrom','TimeOccuredFrom','DateOccuredTo','TimeOccuredTo','datereported','Timereported'], inplace=True)

# Add Weekday -> 0: Monday, 6: Sunday
df2012['Weekday'] = pd.to_datetime(df2012['DateOccuredFrom']).dt.dayofweek.astype(int)
df2013['Weekday'] = pd.to_datetime(df2013['DateOccuredFrom']).dt.dayofweek.astype(int)
df2014['Weekday'] = pd.to_datetime(df2014['DateOccuredFrom']).dt.dayofweek.astype(int)
df2015['Weekday'] = pd.to_datetime(df2015['DateOccuredFrom']).dt.dayofweek.astype(int)
df2016['Weekday'] = pd.to_datetime(df2016['DateOccuredFrom']).dt.dayofweek.astype(int)
df2017['Weekday'] = pd.to_datetime(df2017['DateOccuredFrom']).dt.dayofweek.astype(int)

# Retrieve Hour from Time
df2012['Hour'] = pd.to_datetime(df2012['TimeOccuredFrom']).dt.hour.values.astype(int)
df2013['Hour'] = pd.to_datetime(df2013['TimeOccuredFrom']).dt.hour.values.astype(int)
df2014['Hour'] = pd.to_datetime(df2014['TimeOccuredFrom']).dt.hour.values.astype(int)
df2015['Hour'] = pd.to_datetime(df2015['TimeOccuredFrom']).dt.hour.values.astype(int)
df2016['Hour'] = pd.to_datetime(df2016['TimeOccuredFrom']).dt.hour.values.astype(int)
df2017['Hour'] = pd.to_datetime(df2017['TimeOccuredFrom']).dt.hour.values.astype(int)

# Retrieve Date from Datetime
df2012['Date'] = pd.to_datetime(df2012['DateOccuredFrom']).dt.day.astype(int)
df2013['Date'] = pd.to_datetime(df2013['DateOccuredFrom']).dt.day.astype(int)
df2014['Date'] = pd.to_datetime(df2014['DateOccuredFrom']).dt.day.astype(int)
df2015['Date'] = pd.to_datetime(df2015['DateOccuredFrom']).dt.day.astype(int)
df2016['Date'] = pd.to_datetime(df2016['DateOccuredFrom']).dt.day.astype(int)
df2017['Date'] = pd.to_datetime(df2017['DateOccuredFrom']).dt.day.astype(int)

# Add Time Slot
df2012['Time Slot'] = df2012.apply(time_slot, axis = 1)
df2013['Time Slot'] = df2013.apply(time_slot, axis = 1)
df2014['Time Slot'] = df2014.apply(time_slot, axis = 1)
df2015['Time Slot'] = df2015.apply(time_slot, axis = 1)
df2016['Time Slot'] = df2016.apply(time_slot, axis = 1)
df2017['Time Slot'] = df2017.apply(time_slot, axis = 1)

# Add Week Number 
df2012['Week Number'] = df2012.apply(week_number, axis = 1)
df2013['Week Number'] = df2013.apply(week_number, axis = 1)
df2014['Week Number'] = df2014.apply(week_number, axis = 1)
df2015['Week Number'] = df2015.apply(week_number, axis = 1)
df2016['Week Number'] = df2016.apply(week_number, axis = 1)
df2017['Week Number'] = df2017.apply(week_number, axis = 1)

# Add Week Slot
df2012['Week Slot'] = df2012.apply(week_slot, axis = 1)
df2013['Week Slot'] = df2013.apply(week_slot, axis = 1)
df2014['Week Slot'] = df2014.apply(week_slot, axis = 1)
df2015['Week Slot'] = df2015.apply(week_slot, axis = 1)
df2016['Week Slot'] = df2016.apply(week_slot, axis = 1)
df2017['Week Slot'] = df2017.apply(week_slot, axis = 1)

#Add Season
df2012['Season'] = df2012.apply(season, axis = 1)
df2013['Season'] = df2013.apply(season, axis = 1)
df2014['Season'] = df2014.apply(season, axis = 1)
df2015['Season'] = df2015.apply(season, axis = 1)
df2016['Season'] = df2016.apply(season, axis = 1)
df2017['Season'] = df2017.apply(season, axis = 1)

# Occurence of Crime (DateTime)
df2012['Occurence'] = df2012['DateOccuredTo'] + ' ' + df2012['TimeOccuredTo']
df2013['Occurence'] = df2013['DateOccuredTo'] + ' ' + df2013['TimeOccuredTo']
df2014['Occurence'] = df2014['DateOccuredTo'] + ' ' + df2014['TimeOccuredTo']
df2015['Occurence'] = df2015['DateOccuredTo'] + ' ' + df2015['TimeOccuredTo']
df2016['Occurence'] = df2016['DateOccuredTo'] + ' ' + df2016['TimeOccuredTo']
df2017['Occurence'] = df2017['DateOccuredTo'] + ' ' + df2017['TimeOccuredTo']

# Report of Crime (DateTime)
df2012['Report'] = df2012['datereported'] + ' ' + df2012['Timereported']
df2013['Report'] = df2013['datereported'] + ' ' + df2013['Timereported']
df2014['Report'] = df2014['datereported'] + ' ' + df2014['Timereported']
df2015['Report'] = df2015['datereported'] + ' ' + df2015['Timereported']
df2016['Report'] = df2016['datereported'] + ' ' + df2016['Timereported']
df2017['Report'] = df2017['datereported'] + ' ' + df2017['Timereported']

# Difference between Occurence and Report (Delta)
df2012['Delta time'] = pd.to_datetime(df2012['Report'])- pd.to_datetime(df2012['Occurence'])
df2013['Delta time'] = pd.to_datetime(df2013['Report'])- pd.to_datetime(df2013['Occurence'])
df2014['Delta time'] = pd.to_datetime(df2014['Report'])- pd.to_datetime(df2014['Occurence'])
df2015['Delta time'] = pd.to_datetime(df2015['Report'])- pd.to_datetime(df2015['Occurence'])
df2016['Delta time'] = pd.to_datetime(df2016['Report'])- pd.to_datetime(df2016['Occurence'])
df2017['Delta time'] = pd.to_datetime(df2017['Report'])- pd.to_datetime(df2017['Occurence'])
```

### Common Crimes Across all Years Analysis

```python
def YearWiseCommonAttributeValuesCount(columnname, nameTODisplay):
    arg = columnname
    #Common Major Crime Categories
    cat2012 = df2012.drop_duplicates([arg])[[arg]]
    cat2013 = df2013.drop_duplicates([arg])[[arg]]
    cat2014 = df2014.drop_duplicates([arg])[[arg]]
    cat2015 = df2015.drop_duplicates([arg])[[arg]]
    cat2016 = df2016.drop_duplicates([arg])[[arg]]
    cat2017 = df2017.drop_duplicates([arg])[[arg]]
    crime_cat = cat2012.merge(cat2013).merge(cat2014).merge(cat2015).merge(cat2016).merge(cat2017)
    cat = crime_cat[arg].tolist()
    #Retrieve those crime records in Year 2012,2013,2014,2015,2016,2017,2018
    cat_perc_2012 = df2012.loc[df2012[arg].isin(cat)][[arg]].groupby([arg])[arg].count().reset_index(name="Count")
    cat_perc_2013 = df2013.loc[df2013[arg].isin(cat)][[arg]].groupby([arg])[arg].count().reset_index(name="Count")
    cat_perc_2014 = df2014.loc[df2014[arg].isin(cat)][[arg]].groupby([arg])[arg].count().reset_index(name="Count")
    cat_perc_2015 = df2015.loc[df2015[arg].isin(cat)][[arg]].groupby([arg])[arg].count().reset_index(name="Count")
    cat_perc_2016 = df2016.loc[df2016[arg].isin(cat)][[arg]].groupby([arg])[arg].count().reset_index(name="Count")
    cat_perc_2017 = df2017.loc[df2017[arg].isin(cat)][[arg]].groupby([arg])[arg].count().reset_index(name="Count")
    #Counts of each crime category per year
    df12_15 = pd.merge( pd.merge(pd.merge(cat_perc_2012, cat_perc_2013, on=arg), cat_perc_2014, on=arg),cat_perc_2015, on=arg)
    df12_17 = pd.merge(pd.merge(df12_15, cat_perc_2016, on=arg), cat_perc_2017, on=arg)
    df12_17.columns = [nameTODisplay, 'Year_2012','Year_2013','Year_2014','Year_2015','Year_2016','Year_2017']
    return df12_17
```

#### USC Crime Count Category Analysis

```python
YearWiseCrimeAnalysis = YearWiseCommonAttributeValuesCount('ccClass', 'Crime Category')
YearWiseCrimeAnalysis
```

```python
Year_plot = pd.melt(YearWiseCrimeAnalysis, id_vars=['Crime Category'], var_name=['Year'])
sns.set(style="whitegrid", color_codes=True)
sns.barplot(x="Crime Category", y="value", hue="Year", data=Year_plot,ci="sd",palette="rocket")
sns.set_context("poster")
fig = plt.gcf()
fig.set_size_inches(30, 12)
plt.savefig('USCCrimeCat.png')
plt.ylabel('Crime Count')
plt.xticks(rotation=90)
```

#### Analyze Percentage Difference of each common attribute category over the 2012-2017

```python
def YearwisePercDiff(yearWiseCountData, nameToDisplay):
    yr = yearWiseCountData
    #Percentage Increase/Decrease of each attribute value over 2012-2018
    yr['2012_2013'] = round((yr['Year_2013'] - yr['Year_2012'] ) / yr['Year_2012'] ,2)
    yr['2013_2014'] = round((yr['Year_2014'] - yr['Year_2013'] ) / yr['Year_2013'] ,2) 
    yr['2014_2015'] = round((yr['Year_2015'] - yr['Year_2014'] ) / yr['Year_2014'] ,2)
    yr['2015_2016'] = round((yr['Year_2016'] - yr['Year_2015'] ) / yr['Year_2015'] ,2)
    yr['2016_2017'] = round((yr['Year_2017'] - yr['Year_2016'] ) / yr['Year_2016'] ,2) 
    PercDiff = yr[[nameToDisplay,'2012_2013','2013_2014','2014_2015','2015_2016','2016_2017']]
    return PercDiff

PercCC = YearwisePercDiff(YearWiseCrimeAnalysis, 'Crime Category')
```
